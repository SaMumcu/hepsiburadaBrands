```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
import numpy as np
import numpy
from itertools import zip_longest

alphabeticalUrlParameters=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','diger']
url='https://www.hepsiburada.com/markalar/'
brandNames=[]
brandA=[]
brandB=[]
brandC=[]
brandD=[]
brandE=[]
brandF=[]
brandG=[]
brandH=[]
brandI=[]
brandJ=[]
brandK=[]
brandL=[]
brandM=[]
brandN=[]
brandO=[]
brandP=[]
brandQ=[]
brandR=[]
brandS=[]
brandT=[]
brandU=[]
brandV=[]
brandW=[]
brandX=[]
brandY=[]
brandZ=[]
brandDIGER=[]

 
for parameter in alphabeticalUrlParameters:
    #print(url+parameter)
    page = urlopen(url+parameter)
    document = BeautifulSoup(page, 'html.parser')
    #print(document)
    brandNames = document.find_all('div', attrs={'class': 'brands-area'})
    #print(brandNames)
    for brand in brandNames:
        brand_descendants=brand.descendants
        for descendants in brand_descendants:
            if descendants.name=='span' and descendants.get('class','')==['brand-name'] :
                if parameter==alphabeticalUrlParameters[0]:
                    brandA.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[1]:
                    brandB.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[2]:
                    brandC.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[3]:
                    brandD.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[4]:
                    brandE.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[5]:
                    brandF.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[6]:
                    brandG.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[7]:
                    brandH.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[8]:
                    brandI.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[9]:
                    brandJ.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[10]:
                    brandK.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[11]:
                    brandL.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[12]:
                    brandM.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[13]:
                    brandN.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[14]:
                    brandO.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[15]:
                    brandP.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[16]:
                    brandQ.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[17]:
                    brandR.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[18]:
                    brandS.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[19]:
                    brandT.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[20]:
                    brandU.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[21]:
                    brandV.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[22]:
                    brandW.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[23]:
                    brandX.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[24]:
                    brandY.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[25]:
                    brandZ.append(descendants.text)
                elif parameter==alphabeticalUrlParameters[26]:
                    brandDIGER.append(descendants.text)

resources=[brandA,brandB,brandC,brandD,brandE,brandF,brandG,brandH,brandI,brandJ,brandK,brandL,brandM,brandN,brandO,brandP,brandQ,brandR,brandS,brandT,brandU,brandV,brandW,brandX,brandY,brandZ,brandDIGER]

with open('hbbrands.csv', 'w',newline='',encoding='utf-8') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerows(zip_longest(brandA,brandB,brandC,brandD,brandE,brandF,brandG,brandH,brandI,brandJ,brandK,brandL,brandM,brandN,brandO,brandP,brandQ,brandR,brandS,brandT,brandU,brandV,brandW,brandX,brandY,brandZ,brandDIGER))


print("Writing complete.")


```
